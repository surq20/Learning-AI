# matplotlib实验练习


Matplotlib 是 Python 的绘图库。它可与 NumPy 一起使用，提供了一种有效的 MatLab 开源替代方案，也可以和图形工具包一起使用。

虽然相比其他图形库（Seaborn | pyecharts | plotly | bokeh | pandas_profiling ）这个库丑丑呆呆的，可以说是现在python数据分析中，用的人最多的图形库了。



## 一、导入
### 1.导入matplotlib库简写为plt

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 二、基本图表
### 2.用plot方法画出x=(0,10)间sin的图像

```python
x = np.linspace(0, 10, 30)
plt.plot(x, np.sin(x));
```

### 3.用点加线的方式画出x=(0,10)间sin的图像，提示参数设置为-o

```python
plt.plot(x, np.sin(x), '-o');
```

### 4.用scatter方法画出x=(0,10)间sin的点图像

```python
plt.scatter(x, np.sin(x));
```

### 5.绘制一组误差为±0.8的数据的误差条图，其中行方向x范围为0-10的50个点，列方向为sin(x)
提示： np.linspace，np.random.randn，plt.errorbar

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
 
plt.errorbar(x, y, yerr=dy, fmt='.k')
```

### 6.根据下面数据，绘制一个柱状图,提示：plt.bar

```python
x = [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,9,7,2]
label=['A','B','C','D','E','F','G','H']
 
plt.bar(x,y,tick_label = label);
```

### 7.绘制1000个随机值的直方图

```python
data = np.random.randn(1000)
plt.hist(data);
```

### 8.根据下列数据，在一张图中绘制3组不同的直方图，并设置透明度
提示：kwargs = dict(alpha=0.3, bins=40, density = True)设置透明度

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
 
kwargs = dict(alpha=0.3, bins=40, density = True)
 
plt.hist(x1, **kwargs);
plt.hist(x2, **kwargs);
plt.hist(x3, **kwargs);
```


### 9.绘制x=(0,10)间sin的图像，设置线性为虚线

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),'--');
```

### 10.设置x,y轴标签variable x，value y

```python
x = np.linspace(0.05, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
plt.xlabel('variable x');
plt.ylabel('value y');
```

### 13.在一张图里绘制sin,cos的图形，并展示图例

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
 
ax.plot(x, np.sin(x), label='sin')
ax.plot(x, np.cos(x), '--', label='cos')
ax.legend();
```

### 15.用for创建6个子图

```python
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),fontsize=18, ha='center')
```
