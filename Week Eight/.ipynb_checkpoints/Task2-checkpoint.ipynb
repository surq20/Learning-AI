```python
iris = load_iris()
iris_feature = iris.data
iris_target = iris.target
 
clf = DecisionTreeClassifier()
clf.fit(iris.data, iris.target)
print(clf)
predicted = clf.predict(iris.data)
print(predicted)
print("精度是:{:.3f}".format(clf.score(iris.data, iris.target)))
```
