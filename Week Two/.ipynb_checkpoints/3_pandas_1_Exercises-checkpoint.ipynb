```python
import pandas as pd
```

### Step 2. Import the dataset  student-mat.csv

### Step 3. Assign it to a variable called df.导出上述的csv文件并命名为df,并显示前5行

```python
df = pd.read_csv("student-mat.csv",encoding="gbk")
print(df.head(5))
```

## Step 4. Step 4. For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column将数据集中的列数据从from 'school' until the 'guardian'切片出来，显示前5行查看

```python
stud_alcoh = df.loc[:, "school":"guardian"]
stud_alcoh.head(5)
```

### Step 5. Create a lambda function that capitalize strings.Create a lambda function that capitalize strings.创建一个将字符串大写的lambda函数,提示：capitalizer = lambda x: x.capitalize()

```python
capitalizer = lambda str: str.capitalize()
print(capitalizer("www"))
```

### Step 6. Capitalize both Mjob and Fjob将上面创建的lambda函数应用Mjob and Fjob列中，提示：用apply

```python
stud_alcoh.Mjob.apply(capitalizer)
stud_alcoh.Fjob.apply(capitalizer)
```

### Step 7. Print the last elements of the data set.输出数据集的最后元素，提示：tail()

```python
stud_alcoh.tail()
```

### Step 8. Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.观察上题得到的结果，发现Mjob and Fjob仍然存在没有以大写开头的，请进一步标准化。

```python
stud_alcoh.Mjob = stud_alcoh.Mjob.apply(capitalizer)
stud_alcoh.Fjob = stud_alcoh.Fjob.apply(capitalizer)
stud_alcoh
```

### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)创建一个名为majority的函数（找出年龄大于17岁），将布尔值返回给名为legal_drinker的新列

```python
def majority(age):
    if age > 17:
        return True
    else:
        return False
```

```python
stud_alcoh['legal_drinker'] = stud_alcoh.age.apply(majority)
stud_alcoh
```

### Step 10. Multiply every number of the dataset by 10. 将数据集中整数数据乘以10


```python
def times10(x):
    if type(x) is int:
        return 10 * x
    return x
```

```python
stud_alcoh.applymap(times10).head()
```
