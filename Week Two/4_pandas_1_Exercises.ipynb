# MPG Cars

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset [cars1](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv) and [cars2](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv).  

   ### Step 3. Assign each to a variable called cars1 and cars2,并显示前5行数据

```python
cars1 = pd.read_csv('cars1.csv')
cars2 = pd.read_csv('cars2.csv')
print(cars1.head(5))
print(cars2.head(5))
```

### Step 4. Ops it seems our first dataset has some unnamed blank columns, fix cars1 将数据集cars1 中的空白列去除

```python
cars1 = cars1.loc[:, "mpg" : "car"]
cars1.head()
```

### Step 5. What is the number of observations in each dataset? 统计每个数据集的大小

```python
print(cars1.shape)
print(cars2.shape)
```

### Step 6. Join cars1 and cars2 into a single DataFrame called cars  将cars1和cars2连接到称为cars

```python
cars = cars1.append(cars2)
cars
```

### Step 7. Ops there is a column missing, called owners. Create a random number Series from 15,000 to 73,000. 创建一个从15000到73000的随机数序列的空白列列名为owners。

```python
my_owners = np.random.randint(15000, 73000, 398)
my_owners
```

### Step 8. Add the column owners to cars

```python
cars['owners'] = my_owners
cars.tail()
```
