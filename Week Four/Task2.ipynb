```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
demo = pd.read_csv("avgHgt1.csv")
```

```python
def runplt():
    plt.figure()
    plt.title("Task2")
    plt.xlabel('age')
    plt.ylabel('CHeight')
    plt.grid(True)
    plt.xlim(0, 30)
    plt.ylim(100, 200)
    return plt

xline = demo.loc[:,'age'].values
yline = demo.loc[:,'CHeight'].values
print(xline)
print(yline)
plt = runplt()
plt.plot(xline, yline, 'k.')
plt.show()
```

```python
model = LinearRegression()
X = xline.reshape((-1,1))
y = yline
model.fit(X, y)
plt = runplt()
plt.plot(xline, yline, 'k.')
plt.plot(X, y, 'g-')
plt.show()
```
